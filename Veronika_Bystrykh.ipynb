```python
posts = html.select('div')
for post in posts:
  link = post.select_one('a.b-main__list__link')
  print(link)
```

```python
result_list = []
for post in posts:
  next = post.find_next_sibling('div.b-main')
  title = post.select_one('h2.b-news__short__title')
  text = post.select_one('div.b-news__short__text')
  score = post.select_one('span.b-news__short__view')
  category = post.select_one('span.b-news__short__category')
  result_list.append({
    'title': title,
    'text': text,
    'category': category,
    'score': score,
  })
result_list
```

```python
import pandas as pd
table = pd.DataFrame(result_list).convert_dtypes()
table
```
